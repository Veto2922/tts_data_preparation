```python
import yt_dlp

channel_url = "https://www.youtube.com/@CHANNEL_NAME/videos"  # غيرها باسم القناة

ydl_opts = {
    'outtmpl': '%(title)s.%(ext)s',   # اسم الملف = عنوان الفيديو
    'format': 'bestvideo+bestaudio/best',  # أعلى جودة
    'merge_output_format': 'mp4',
    'ignoreerrors': True,  # يكمل حتى لو في فيديو فيه مشكلة
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([channel_url])
```
